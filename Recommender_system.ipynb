```python
#importing necessary libraries
import numpy as np
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel
#putting electronics data on 'ds' dataframe
ds = pd.read_csv('catalog_products.csv')

tfidf = TfidfVectorizer(stop_words='english')
ds['collection'] = ds['collection'].fillna('')
#Construct the required TF-IDF matrix by applying the fit_transform method on the Category1 feature
category_matrix = tfidf.fit_transform(ds['collection'])
#Output the shape of tfidf_matrix
category_matrix.shape

similarity_matrix = linear_kernel(category_matrix,category_matrix)
similarity_matrix

#items index mapping
mapping = pd.Series(ds.index,index = ds['name'])
mapping

def recommend_items_based_on_category(item_input):
 
    item_index = mapping[item_input][0]
    #get similarity values with other items
    #similarity_score is the list of index and similarity matrix
    similarity_score = list(enumerate(similarity_matrix[item_index]))
    #sort in descending order the similarity score of item inputted with all the other items
    similarity_score = sorted(similarity_score, key=lambda x: x[1], reverse=True)
    # Get the scores of the 20 most similar items. Ignore the first item.
    similarity_score = similarity_score[1:21]
    #return item names using the mapping series
    item_indices = [i[0] for i in similarity_score]
    return (ds.iloc[item_indices])

def get_price(item_input):
    item_index = mapping[item_input][0]
    return (ds['price'].iloc[item_index])
similar_items = recommend_items_based_on_category('Alpha a5100 Mirrorless Digital Camera with 16-50mm Lens (White)')
#print(similar_items)
itemprice = get_price('Alpha a5100 Mirrorless Digital Camera with 16-50mm Lens (White)')
#lowprice = lowprice[0:0]
x = 0

for i in range(len(similar_items)):
    if similar_items['price'].iloc[i] < itemprice:
            print(similar_items['price'].iloc[i])
            print(similar_items['name'].iloc[i])
            
        #lowitem = similar_items['price'].iloc[i]
        #lowprice = pd.concat([lowprice, lowitem])
        #print(lowitem)
        #for x in highprice:
            #similar_items.drop(similar_items[similar_items.price == x])
       #print(similar_items['name'].iloc[i])
#print(lowprice['name'])
#names = lowprice['name'].head(10)
#prices = lowprice['price'].head(10)
#category = lowprice['Category1'].head(10)
#print(names)
#print(prices)
#print(category)

#for i in similar_items:
    #if similar_items['price'] < itemprice:
        #print(similar_items['name'] + "Price:" + similar_items['price'].to_string())
```
